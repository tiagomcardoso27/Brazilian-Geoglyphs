```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

```python
def Calc_Percentile(scores, percentile_rank):
    scores.sort()
    percentiles = [stats.percentileofscore(scores, a, 'weak') for a in scores]
    for i,score in enumerate(scores):
        if percentiles[i] >= percentile_rank:
            return score
```

```python
edges = cv2.Canny(sub_binary,5,50)
lines = cv2.HoughLines(edges,1,np.pi/180,80)
for lin in lines:
    #rho = lin[0][0]
    #theta = lin[0][1]
    rho,theta = lin[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))

    cv2.line(sub_binary,(x1,y1),(x2,y2),(128,255,0),2)
lines.shape
```

```python
#Problema: theta = 0.7853982 / 2.3561945
copy = geo
size = np.size(sub_binary)
skel = np.zeros(sub_binary.shape,np.uint8)
 
ret,sub_binary = cv2.threshold(sub_binary,100,255,0)
element = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
done = False
 
while( not done):
    eroded = cv2.erode(sub_binary,element)
    temp = cv2.dilate(eroded,element)
    temp = cv2.subtract(sub_binary,temp)
    skel = cv2.bitwise_or(skel,temp)
    sub_binary = eroded.copy()
 
    zeros = size - cv2.countNonZero(sub_binary)
    if zeros==size:
        done = True
        
edges = cv2.Canny(skel,5,50)
lines = cv2.HoughLines(edges,1,np.pi/180,170)
for lin in lines:
    #rho = lin[0][0]
    #theta = lin[0][1]
    rho,theta = lin[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))

    cv2.line(copy,(x1,y1),(x2,y2),(128,255,0),2)
lines.shape
```

```python
circles = cv2.HoughCircles(sub_binary, cv2.HOUGH_GRADIENT, dp=1.5, minDist=300)
if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    
    for (x, y, r) in circles:
        cv2.circle(sub_binary, (x, y), r, (128, 255, 0), 4)
print(circles)
```
